```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

categorical = ['type', 'group', 'education', 'meal', 'preparation course']


column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(), categorical),
])

df_train_transformed = column_transformer.fit_transform(df_train)
df_test_transformed = column_transformer.transform(df_test)
```

```python
df_train_transformed.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
from sklearn.metrics import roc_auc_score

y = df_train['Pass']
tree = DecisionTreeClassifier(max_depth=5, criterion='entropy')
res = cross_val_score(tree, df_train_transformed, y, cv=3, scoring='roc_auc').mean()
```

```python
from catboost import CatBoostClassifier
import numpy as np

catboost = CatBoostClassifier(eval_metric='AUC')

params = {'max_depth': np.arange(1, 15, 2),
          'n_estimators': np.arange(50, 251, 50),
          'learning_rate': np.arange(1e-4, 1e-1, 5),
          'subsample':  np.arange(0.1, 1, 0.2),
          'early_stopping_rounds': np.arange(10, 100, 20)}

grid_search_result = catboost.grid_search(params, df_train_transformed, y, verbose=False)
```

```python
coeffs = grid_search_result['params']
print(coeffs)
```

```python
catboost_model = CatBoostClassifier(eval_metric='AUC',
                                    l2_leaf_reg=coeffs['od_wait'],
                                    learning_rate=coeffs['learning_rate'],
                                    max_depth=coeffs['depth'],
                                    subsample=coeffs['subsample'],
                                    n_estimators=coeffs['iterations'])
catboost_model.fit(df_train_transformed, y, verbose=False)

print(catboost_model.predict(df_test_transformed))
# print(f'Catboost score: {roc_auc_score(y_test == 1, catboost_model.predict_proba(x_test)[:, 1]):.4f}')
```
